```python
import numpy as np
T = np.pi
n = np.linspace(-T,T,1000)
```

##a.) sin[𝑛]

```python
first_dis = np.abs(np.sin(n)) @ np.abs(np.sin(n))
p = (1/2*T)*first_dis
print("first_dis:",first_dis)
print (p,"Discrete")
```

##b.) cos(𝑡)

```python
first_cos = (1/2)*n + 1/4*np.sin(2*n)
pb = (1/2*T)*first_cos
pb
```

d. $y[n] = \frac{2}{\sqrt{\pi}}{e}^{-lπ}$

###Part 2. [100 pts] Prove the equality of the following trigonometric functions using Euler identities. Further prove the equality in Python. Furthermore, plot the left-hand side and right-hand side of the equation and their Euler forms as additional proof

### $\sin^2(\theta) = \frac{1}{2}(1 - \cos(2\theta))$

####LEFT SIDE

```python
import matplotlib.pyplot as plt
left_side = np.sin(n) ** 2
plt.plot(n,left_side, label=r'$\frac{1}{2}(1-\cos2\theta)$')
plt.title(r'Plot of $\frac{1}{2}(1-\cos2\theta)$')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\sin^2(\theta)$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
right_side = ((1/2) * (1-np.cos(2*n)))
plt.plot(n,right_side, label=r'$\sin^2(\theta)$')
plt.title(r'Plot of $\sin^2(\theta)$')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\sin^2(\theta)$')
plt.legend()
plt.grid(True)
plt.show()
```
