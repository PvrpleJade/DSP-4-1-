```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv("/content/ArevaloHouseholdLasPinas.csv")
data2 = pd.read_csv("/content/ColumbaHouseholdCaloocan.csv")
```

```python
data1.head()
```

```python
import statistics
print (data1)
Temperature = data1['Temp'].tolist()

mean_value = statistics.mean(Temperature)

print("Mean:", mean_value)
```

```python
import statistics
print (data2)
Temperature = data2['Temp'].tolist()

mean_value = statistics.mean(Temperature)
median_value = statistics.median(Temperature)
mode_value = statistics.mode(Temperature)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```
